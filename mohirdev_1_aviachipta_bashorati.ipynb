#Train datani mlga tayyorlaymiz

One hot encoder yordamida matnli ustunlarni sonlarga o'zgartiramiz faqat 'flight' ustunidan boshqa chunki unda unikal qiymatlar juda ko'p va mantiqan uning chipta narxiga ta'siri ko'p emas.

```python
train_data = train_data.drop('flight', axis=1)
train_data.head()
```

```python
#Raqamli ustunlar uchun pipiline
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

num_pipeline = Pipeline([
          ('std_scaler', StandardScaler())
])
```

```python
#full pipeline
from sklearn.compose import ColumnTransformer

num_attribs = ['duration', 'days_left']
cat_attribs = ['airline', 'source_city', 'departure_time', 'stops',	'arrival_time',	'destination_city',	'class']

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attribs),
    ('cat', OneHotEncoder(), cat_attribs)
])
```

```python
X = train_data.drop('price', axis=1)
y = train_data['price'].copy()

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.01, random_state=42)
```

```python
X_train_prepared = full_pipeline.fit_transform(X_train)
```

```python
X_train_prepared.shape
```

#Model qurishni boshlaymiz

Random Forestni sinab ko'ramiz

```python
from sklearn.ensemble import RandomForestRegressor

RF_model = RandomForestRegressor(n_estimators=100)
RF_model.fit(X_train_prepared, y_train)
```

```python
X_test_prepared = full_pipeline.fit_transform(X_test)
```

```python
y_predicted = RF_model.predict(X_test_prepared)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

MAE = mean_absolute_error(y_test, y_predicted)
RMSE = np.sqrt(mean_squared_error(y_test, y_predicted))
print(f'O\'rtacha absolyut xatolik: {MAE}')
print(f'O\'rtacha kvadratik xatolik: {RMSE}')
```

XGBoostni sinaymiz

```python
import xgboost as xg

xgb_model = xg.XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.05, max_depth=15)
xgb_model.fit(X_train_prepared, y_train)
```

```python
y_predicted_xgb = xgb_model.predict(X_test_prepared)
```

```python
MAE = mean_absolute_error(y_test, y_predicted_xgb)
RMSE = np.sqrt(mean_squared_error(y_test, y_predicted_xgb))
print(f'O\'rtacha absolyut xatolik: {MAE}')
print(f'O\'rtacha kvadratik xatolik: {RMSE}')
```

Decision treeni sinaymiz

```python
from sklearn.tree import DecisionTreeRegressor
DT_model = DecisionTreeRegressor(max_depth=5)
DT_model.fit(X_train_prepared, y_train)
```

```python
y_predicted_dt = DT_model.predict(X_test_prepared)

MAE = mean_absolute_error(y_test, y_predicted_dt)
RMSE = np.sqrt(mean_squared_error(y_test, y_predicted_dt))
print(f'O\'rtacha absolyut xatolik: {MAE}')
print(f'O\'rtacha kvadratik xatolik: {RMSE}')
```
