Klassik dastur

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt

# data
data = make_regression(n_samples=55, n_features=1, noise=8, random_state=32)
x_data, y_data = data[0], data[1]



def classic_program(x): return x*70

y_pred = [classic_program(x) for x in x_data]

# plotting
plt.scatter(x_data, y_data, label='Haqiqiy data sets')
plt.plot(x_data, y_pred, 'r', label='Dastur')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('x*70')
```

Vazn qo'llanilgan

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt

# data
data = make_regression(n_samples=55, n_features=1, noise=8, random_state=32)
x_data, y_data = data[0], data[1]



def classic_program(x, w): return x*w*70

w = 0.9

y_pred = [classic_program(x, w) for x in x_data]

# plotting
plt.scatter(x_data, y_data, label='Haqiqiy data sets')
plt.plot(x_data, y_pred, 'r', label='Dastur')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title(f'{w}*x*70')
```

ML dastur

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# data
data = make_regression(n_samples=55, n_features=1, noise=8, random_state=32)
x_data, y_data = data[0], data[1]



LR = LinearRegression().fit(x_data, y_data)
y_pred = LR.predict(x_data)

equation = f'{LR.coef_[0]:.3f}*x+{LR.intercept_:.3f}'
# plotting
plt.scatter(x_data, y_data, label='Haqiqiy data sets')
plt.plot(x_data, y_pred, 'r', label='ML Dastur')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title(equation)
```

DL dastur

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor

# data
data = make_regression(n_samples=55, n_features=1, noise=8, random_state=32)
x_data, y_data = data[0], data[1]



LR = MLPRegressor(max_iter=2000).fit(x_data, y_data)
y_pred = LR.predict(x_data)


# plotting
plt.scatter(x_data, y_data, label='Haqiqiy data sets')
plt.plot(x_data, y_pred, 'r', label='DL Dastur')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```
